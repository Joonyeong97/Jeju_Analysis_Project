```python
clementine =pd.read_csv("C:/workR/1-project/ppt/지영 _감귤 생산량/cle_10.csv", 
                        encoding='UTF-8')

clementine.index=clementine['연도']
# drop axis의 경우 column이면 1, row이면 0이다.

clementine.drop(labels='연도', axis=1, inplace=True) 
cle=clementine.iloc[:,1:]
```
