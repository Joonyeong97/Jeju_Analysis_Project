```python
weatherBloom =pd.read_csv("C:/Users/ICT01_10/Documents/Gina/1-project/j4,5.csv", 
                        encoding='CP949')

weatherBloom.columns=['Bloom', 'TempOfBloom', 'PrecipitationOfBloom', 'InsolationOfBloom', 
                     'HumidityOfBloom', 'WindspeedOfBloom']
weatherBloom.drop(labels="Bloom", axis=1, inplace=True)
weatherBloom
```

```python
y=mandarinDataset.iloc[:,-1]
X=mandarinDataset.iloc[:,1:]

plt.figure(figsize=(15,15))
sns.heatmap(data=X.corr(), annot=True, fmt='.3f', linewidths=.5, cmap='hot_r')
```

```python
from sklearn.feature_selection import RFE #recursive feature elimination 
from sklearn.svm import SVR # 이번엔 분류! SVRegression

estimator =SVR(kernel='linear') #선형회귀Regression 예측 

# 분산이 작은 데이터 == 가장 영향력 없는 변수 > noise 제거 noise elimination)
selector =RFE(estimator, 5, step=1) # step=1 :1번에 1개씩 제거해라
# 5개 변수에 대해서 스텝 1개씩. = 10개 변수중에서 5개만 남기겠다. 

# RFE(): ( 기존의 추정기를 받고 , 변수추출갯수(제거후), step 지정 )
selector = selector.fit(X, y)
selector.support_
```

```python
selector.ranking_
```

```python
selector.n_features_ # 잡음제거한 변수 5개 남김.
```

```python
selector.estimator_ # 변수를 제거한 추정기.
```

# PCA 모델 입력 전단계 = 특징추출하기 = noise 제거

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
print(pca.explained_variance_ratio_)

print(pca.explained_variance_) # 분산이 큰것이 주성분~~!
print(pca.noise_variance_) # Noise
```
