```python
from sklearn.feature_selection import RFE #recursive feature elimination 
from sklearn.svm import SVR # 이번엔 분류! SVRegression

estimator =SVR(kernel='linear') #선형회귀Regression 예측 

# 분산이 작은 데이터 == 가장 영향력 없는 변수 > noise 제거 noise elimination)
selector =RFE(estimator, 5, step=1) # step=1 :1번에 1개씩 제거해라
# 5개 변수에 대해서 스텝 1개씩. = 10개 변수중에서 5개만 남기겠다. 

# RFE(): ( 기존의 추정기를 받고 , 변수추출갯수(제거후), step 지정 )
selector = selector.fit(X, y)
selector.support_
```

# PCA 모델 입력 전단계 = 특징추출하기 = noise 제거

```python
# PHU 에 상관관계가 큰 개화기 수확기 평균기온 제거 및, 다중공선성이 보였던, 수확기 일조량 제거된 데이터셋
mandarin_extraction=mandarinDataset[["productionT","WindspeedOfBloom","PHUOfBloom","HumidityOfHarvest","WindspeedOfHarvest","PHUOfHarvest"]]
mandarin_extraction
```

```python
# 정규화 4가지 버젼 다 해보쟝
from sklearn.preprocessing import scale, robust_scale, minmax_scale, maxabs_scale, StandardScaler

y=mandarin_extraction.iloc[:,0]
X=mandarin_extraction.iloc[:,1:]

scale_X = scale(X)
robust_X = robust_scale(X)
minmax_X = minmax_scale(X)
maxabs_X = maxabs_scale(X)

scaler = StandardScaler() # 표준정규분포기에 데이터 넣기위해 인스턴스
stdScale_X = scaler.fit(X)
std_transformed_X = scaler.transform(X)
X.std(), std_transformed_X.std()
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scale_X, y, random_state=0)


# 최소자승법을 해봅시다!
import statsmodels.api as sm
result = sm.OLS(y_train, X_train).fit()
result.summary()

# R-squared => 설명력 : 회귀제곱합/ 총오차제곱합 0.001
# Adj. R-squared 모델을 통해 수정된 값을 쓴다 -0.054  정규분포로 안바꿔주니 못쓰는값...
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(robust_X, y, random_state=0)



import statsmodels.api as sm
result = sm.OLS(y_train, X_train).fit()
result.summary()


# R-squared => 설명력 : 회귀제곱합/ 총오차제곱합 0.001  =>0.021
# Adj. R-squared 모델을 통해 수정된 값을 쓴다 -0.054  robust 정규분포로 =>-0.034
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(minmax_X , y, random_state=0)


# 최소자승법을 해봅시다!
import statsmodels.api as sm
result = sm.OLS(y_train, X_train).fit()
result.summary()


# R-squared => 설명력 : 회귀제곱합/ 총오차제곱합 0.001  =>0.885
# Adj. R-squared 모델을 통해 수정된 값을 쓴다 -0.054   minmax 정규분포로 =>0.878
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(maxabs_X , y, random_state=0)


# 최소자승법을 해봅시다!
import statsmodels.api as sm
result = sm.OLS(y_train, X_train).fit()
result.summary()


# R-squared => 설명력 : 회귀제곱합/ 총오차제곱합 0.001  =>0.993
# Adj. R-squared 모델을 통해 수정된 값을 쓴다 -0.054   maxabs 정규분포로 =>0.993
```
